```python
def remove_stopwords(token_list):
    """
    Retire les stop words et conserve uniquement les tokens qui contiennent au moins une lettre ou un chiffre.
    """
    stop_words = nlp.Defaults.stop_words
    
    pattern = re.compile(r'.*[A-Za-z0-9].*')
    
    cleaned_tokens = [
        token for token in token_list
        if token.lower() not in stop_words and pattern.match(token)
    ]

    lower_tokens = [token.lower() for token in cleaned_tokens]
        
    return lower_tokens

tokens = ['-', 'Toi-meme', '.', '1.5', '1,2', '#', '#world', '#1.5']

cleaned_tokens = remove_stopwords(tokens)
print("Tokens sans stop words :", cleaned_tokens)
```

```python
def remove_stopwords_and_lemmatize(token_list):
    """
    Retire les stop words et conserve uniquement les tokens contenant au moins une lettre ou un chiffre.
    Applique ensuite la lemmatization pour obtenir la forme canonique de chaque token.
    """
    stop_words = nlp.Defaults.stop_words
    # On garde les tokens qui contiennent au moins un caractère alphanumérique
    pattern = re.compile(r'.*[A-Za-z0-9].*')
    
    # Filtrer les tokens et les convertir en minuscules
    cleaned_tokens = [
        token.lower() for token in token_list 
        if token.lower() not in stop_words and pattern.match(token)
    ]
    
    # Traiter la liste des tokens avec spaCy
    # Ici, on reconstruit une chaîne de caractères à partir des tokens pour que spaCy les re-tokenize
    doc = nlp(" ".join(cleaned_tokens))
    
    # Extraire la lemmatisation de chaque token dans le doc
    lemmatized_tokens = [token.lemma_ for token in doc]
    
    return lemmatized_tokens

# Exemple de tokens
tokens = ["organizations", "organization", "organizing", "organized", "I", "or", 'to']

cleaned_tokens = remove_stopwords_and_lemmatize(tokens)
print("Tokens après le filtrage et la lemmatization :", cleaned_tokens)
```

```python
def create_sentence_dict(example, dataset):
    """
    Convertit un exemple brut du dataset HuggingFace en liste de dictionnaires
    contenant les tokens et leurs tags en texte.
    """
    tokens = example['tokens']
    pos_ids = example['pos_tags']
    chunk_ids = example['chunk_tags']
    ner_ids = example['ner_tags']

    # On utilise les mappings fournis par HuggingFace
    pos_tags = [dataset.features['pos_tags'].feature.names[idx] for idx in pos_ids]
    chunk_tags = [dataset.features['chunk_tags'].feature.names[idx] for idx in chunk_ids]
    ner_tags = [dataset.features['ner_tags'].feature.names[idx] for idx in ner_ids]

    # On crée la liste de dictionnaires
    sentence = []
    for token, pos, chunk, ner in zip(tokens, pos_tags, chunk_tags, ner_tags):
        sentence.append({
            'tokens': token,
            'pos_tags': pos,
            'chunk_tags': chunk,
            'ner_tags': ner
        })

    return sentence
```

```python
import re
import spacy
import sklearn_crfsuite
from sklearn_crfsuite import metrics
from datasets import load_dataset
from tqdm import tqdm

# Charger le modèle spaCy (pour la lemmatisation)
nlp = spacy.load("en_core_web_sm")

# --- Prétraitement : retirer les stop words et appliquer la lemmatisation sans transformer en minuscule ---
def remove_stopwords_and_lemmatize(token_list):
    """
    Retire les stop words et conserve uniquement les tokens contenant au moins une lettre ou un chiffre.
    Applique ensuite la lemmatisation pour obtenir la forme canonique de chaque token.
    La casse originale est préservée.
    """
    stop_words = nlp.Defaults.stop_words
    pattern = re.compile(r'.*[A-Za-z0-9].*')
    
    # On vérifie en lower pour le filtrage, mais on conserve le token original
    cleaned_tokens = [
        token for token in token_list 
        if token.lower() not in stop_words and pattern.match(token)
    ]
    
    # Reconstruire une chaîne pour spaCy sans changer la casse
    doc = nlp(" ".join(cleaned_tokens))
    lemmatized_tokens = [token.lemma_ for token in doc]
    return lemmatized_tokens

# Exemple d'utilisation du prétraitement
tokens_example = ["organizations", "organization", "organizing", "organized", "I", "or", "to"]
print("Tokens après le filtrage et la lemmatization :", remove_stopwords_and_lemmatize(tokens_example))

# --- Chargement du dataset CoNLL-2003 ---
dataset = load_dataset("conll2003")
train_dataset = dataset['train']
test_dataset = dataset['test']

# --- Conversion d'un exemple brut en phrase formatée ---
def create_sentence_dict(example, ds):
    tokens = example['tokens']
    pos_ids = example['pos_tags']
    chunk_ids = example['chunk_tags']
    ner_ids = example['ner_tags']
    
    # Mapper les IDs vers leurs labels textuels
    pos_tags = [ds.features['pos_tags'].feature.names[idx] for idx in pos_ids]
    chunk_tags = [ds.features['chunk_tags'].feature.names[idx] for idx in chunk_ids]
    ner_tags = [ds.features['ner_tags'].feature.names[idx] for idx in ner_ids]
    
    sentence = []
    for token, pos, chunk, ner in zip(tokens, pos_tags, chunk_tags, ner_tags):
        sentence.append({
            'tokens': token,
            'pos_tags': pos,
            'chunk_tags': chunk,
            'ner_tags': ner
        })
    return sentence

# --- Extraction des features pour chaque token dans une phrase ---
def word2features(sentence, i):
    word = sentence[i]['tokens']
    postag = sentence[i]['pos_tags']
    chunktag = sentence[i]['chunk_tags']
    
    features = {
        'bias': 1.0,
        'word.lower()': word.lower(),  # utile pour normaliser certaines features
        'word[-3:]': word[-3:],
        'word[-2:]': word[-2:],
        'word.isupper()': word.isupper(),
        'word.istitle()': word.istitle(),
        'word.isdigit()': word.isdigit(),
        # POS features
        'postag': postag,
        'postag[:2]': postag[:2],
        'is_noun': postag in ['NN', 'NNP', 'NNS', 'NNPS'],
        'is_proper_noun': postag == 'NNP',
        'is_number': postag == 'CD',
        'is_verb': postag in ['VB', 'VBD', 'VBG', 'VBN', 'VBP', 'VBZ'],
        # Chunk features
        'chunktag': chunktag,
        'in_np_chunk': chunktag in ['B-NP', 'I-NP'],
        'in_vp_chunk': chunktag in ['B-VP', 'I-VP'],
        'in_pp_chunk': chunktag in ['B-PP', 'I-PP'],
        'is_outside_chunk': chunktag == 'O',
        # Autres
        'word_length': len(word),
    }
    
    # Contexte précédent
    if i > 0:
        prev_word = sentence[i-1]['tokens']
        prev_postag = sentence[i-1]['pos_tags']
        prev_chunktag = sentence[i-1]['chunk_tags']
        features.update({
            '-1:word.lower()': prev_word.lower(),
            '-1:word.istitle()': prev_word.istitle(),
            '-1:word.isupper()': prev_word.isupper(),
            '-1:postag': prev_postag,
            '-1:postag[:2]': prev_postag[:2],
            '-1:chunktag': prev_chunktag,
        })
    else:
        features['BOS'] = True  # Beginning Of Sentence
    
    # Contexte suivant
    if i < len(sentence) - 1:
        next_word = sentence[i+1]['tokens']
        next_postag = sentence[i+1]['pos_tags']
        next_chunktag = sentence[i+1]['chunk_tags']
        features.update({
            '+1:word.lower()': next_word.lower(),
            '+1:word.istitle()': next_word.istitle(),
            '+1:word.isupper()': next_word.isupper(),
            '+1:postag': next_postag,
            '+1:postag[:2]': next_postag[:2],
            '+1:chunktag': next_chunktag,
        })
    else:
        features['EOS'] = True  # End Of Sentence
    
    return features

# Fonctions pour transformer une phrase en features et labels
def sent2features(sentence):
    return [word2features(sentence, i) for i in range(len(sentence))]

def sent2labels(sentence):
    return [word['ner_tags'] for word in sentence]
```

```python
# --- Conversion du dataset train et test en phrases formatées ---
train_sentences = [create_sentence_dict(example, train_dataset) for example in tqdm(train_dataset, desc="Converting train dataset")]
test_sentences = [create_sentence_dict(example, test_dataset) for example in tqdm(test_dataset, desc="Converting test dataset")]

# --- Extraction des features et des labels ---
X_train = [sent2features(sentence) for sentence in tqdm(train_sentences, desc="Extracting features for train")]
y_train = [sent2labels(sentence) for sentence in tqdm(train_sentences, desc="Extracting labels for train")]

X_test = [sent2features(sentence) for sentence in tqdm(test_sentences, desc="Extracting features for test")]
y_test = [sent2labels(sentence) for sentence in tqdm(test_sentences, desc="Extracting labels for test")]

# --- Définition et entraînement du modèle CRF ---
crf = sklearn_crfsuite.CRF(
    algorithm='lbfgs',
    c1=0.1,
    c2=0.1,
    max_iterations=100,
    all_possible_transitions=False
)

print("Training CRF model...")
crf.fit(X_train, y_train)

# --- Prédiction et évaluation sur le jeu de test ---
print("Predicting on test set...")
y_pred = crf.predict(X_test)
print(metrics.flat_classification_report(y_test, y_pred))
```

```python
import spacy
from rdflib import Graph, URIRef, Literal, Namespace
from rdflib.namespace import RDF
from tqdm import tqdm

# Charger spaCy
nlp = spacy.load("en_core_web_sm")

# Notre texte
text = """Star Wars IV is a Movie where there are different kinds of creatures, like humans and wookies.
Some creatures are Jedis; for instance, the human Luke is a Jedi, and Master Yoda – for whom the species is not known – is also a Jedi.
The wookie named Chewbacca is Han’s co-pilot on the Millennium Falcon starship.
The speed of Millennium Falcon is 1.5 (above the speed of light!)
"""

# Traiter le texte
doc = nlp(text)

# Initialiser un graphe RDF
g = Graph()

# Définir le namespace
EX = Namespace("http://example.org/")
g.bind("ex", EX)

# Fonction pour rendre les noms URI friendly
def to_uri(name):
    return URIRef(EX[str(name).replace(" ", "_").replace("’", "").replace("-", "_")])

# Extraire des relations simples (subject, predicate, object)
relations = []

for sent in tqdm(doc.sents, desc="Extraction des relations"):
    parsed_sent = nlp(sent.text)
    
    subject = None
    predicate = None
    obj = None
    
    for token in parsed_sent:
        # Chercher un sujet
        if token.dep_ in ["nsubj", "nsubjpass"]:
            subject = token.text
            
            # Le verbe ou prédicat est la head (racine)
            predicate = token.head.lemma_
            
            # Chercher les objets de ce verbe (dobj ou attr pour les attributs)
            for child in token.head.children:
                if child.dep_ in ["dobj", "attr", "prep", "pobj"]:
                    obj = child.text
                    
                    # On a un triplet
                    if subject and predicate and obj:
                        relations.append((subject, predicate, obj))

# Afficher les relations extraites
print("Relations extraites :\n")
for r in relations:
    print(f"{r[0]} -- {r[1]} --> {r[2]}")

# Ajouter ces relations au graphe RDF
for (subj, pred, obj) in relations:
    subj_uri = to_uri(subj)
    pred_uri = to_uri(pred)
    obj_uri = to_uri(obj)
    
    # Ajout du triplet RDF
    g.add((subj_uri, pred_uri, obj_uri))

# Sérialiser le graphe RDF en XML
print("\nRDF/XML serialization :\n")
print(g.serialize(format="xml"))

# Exécuter une requête SPARQL simple sur le graphe
query = """
SELECT ?subject ?predicate ?object
WHERE {
  ?subject ?predicate ?object .
}
"""

print("\nRésultat de la requête SPARQL :\n")
for row in g.query(query):
    print(f"{row.subject} {row.predicate} {row.object}")
```
