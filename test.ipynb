```python
import requests

url = "https://fr.wikipedia.org/wiki/Oscar_Jespers"
response = requests.get(url)
html_content = response.text  # contenu HTML renvoyé par la page
```

```python
print(html_content)
```

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(html_content, "html.parser")

# Exemple : extraire tous les titres de niveau h1
titles = soup.find_all("h2")
for t in titles:
    print(t.text)


```

```python
title_list = []
max_level = 6  # Limite aux niveaux h1 à h6

for i in range(1, max_level + 1):
    str_title = f"h{i}"
    titles = soup.find_all(str_title)

    if not titles:
        continue  # Passe au prochain niveau s'il n'y a pas de titre

    for t in titles:
        title_list.append(t.get_text(strip=True))  # Ajoute le texte sans espaces inutiles
        t.decompose()  # Supprime la balise du DOM après l'avoir traitée
```

```python
print(title_list)
```
